```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Directorio donde están los archivos
data_dir = r"C:\Users\Mariana\Documents\andros_data_processing\YSI_profiles\Priority_Deep"

# Lista de archivos CSV (puedes ordenarlos manualmente si es necesario)
file_names = [
    "AW5D_YSI_20250225.csv",
    "AW6D_YSI_20250226.csv",
    "BW3_YSI_20250309_10.csv",
    "LRS69D_YSI_20250222R.csv",
    "LRS70D_YSI_20250317.csv"
]

# Construir rutas completas
file_paths = [os.path.join(data_dir, name) for name in file_names]

# Leer los datos
data = [pd.read_csv(file) for file in file_paths]

# Determinar límites comunes
x_min = min(df["SpCond_muS/cm"].min() for df in data)
x_max = max(df["SpCond_muS/cm"].max() for df in data)
y_min = min(df["Vertical Position m"].min() for df in data)
y_max = max(df["Vertical Position m"].max() for df in data)

# -------- Figura 1: Escala lineal --------
fig1, axes1 = plt.subplots(1, 5, figsize=(22, 6), sharey=True)
for i, ax in enumerate(axes1):
    df = data[i]
    ax.plot(df["SpCond_muS/cm"], df["Vertical Position m"])
    ax.set_title(file_names[i], fontsize=9)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_max, y_min)
    ax.set_xlabel("SpCond (μS/cm)")
    if i == 0:
        ax.set_ylabel("Vertical Position (m)")
fig1.suptitle("Conductividad específica (Escala Lineal)", fontsize=14)
plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()

# -------- Figura 2: Escala logarítmica --------
fig2, axes2 = plt.subplots(1, 5, figsize=(22, 6), sharey=True)
for i, ax in enumerate(axes2):
    df = data[i]
    ax.plot(df["SpCond_muS/cm"], df["Vertical Position m"])
    ax.set_title(file_names[i], fontsize=9)
    ax.set_xscale("log")
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_max, y_min)
    ax.set_xlabel("SpCond (μS/cm) [log]")
    if i == 0:
        ax.set_ylabel("Vertical Position (m)")
fig2.suptitle("Conductividad específica (Escala Logarítmica)", fontsize=14)
plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()
```
